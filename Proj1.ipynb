```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Select features and target
X = df['MedianIncome'].values.reshape(-1, 1)  # Selecting a single feature for simplicity
y = df['MedianHouseValue'].values

# Add intercept term to X
X_b = np.c_[np.ones((len(X), 1)), X]

# Gradient Descent parameters
learning_rate = 0.01
epochs = 1000

# Initialize parameters (theta)
theta = np.random.randn(2, 1)  # Initialize with random values

# Define the cost function (Mean Squared Error)
def compute_cost(X, y, theta):
    predictions = np.dot(X, theta)
    mse = np.mean((predictions - y.reshape(-1, 1)) ** 2)
    return mse

# Gradient Descent function to update parameters
def gradient_descent(X, y, theta, learning_rate, epochs):
    m = len(y)
    costs = []
    for epoch in range(epochs):
        # Compute predictions and error
        predictions = np.dot(X, theta)
        error = predictions - y.reshape(-1, 1)
        
        # Compute gradients
        gradient = (2/m) * np.dot(X.T, error)
        
        # Update parameters
        theta -= learning_rate * gradient
        
        # Compute and store cost
        cost = compute_cost(X, y, theta)
        costs.append(cost)
        
        # Print progress
        if epoch % 100 == 0:
            print(f"Epoch {epoch}: Cost = {cost}")
    
    return theta, costs

# Perform Gradient Descent
theta_best, costs = gradient_descent(X_b, y, theta, learning_rate, epochs)

# Print final parameters and cost
print("\nFinal Parameters (theta):")
print(theta_best)
print(f"Final Cost: {costs[-1]}")

# Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, X_b.dot(theta_best), color='red', label='Best Fit Line')
plt.title('Best Fit Line using Gradient Descent')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Select features and target
X = df['MedianIncome'].values.reshape(-1, 1)  # Selecting a single feature for simplicity
y = df['MedianHouseValue'].values

# Add intercept term to X
X_b = np.c_[np.ones((len(X), 1)), X]

# Stochastic Gradient Descent parameters
learning_rate = 0.01
epochs = 1000

# Initialize parameters (theta)
theta = np.random.randn(2, 1)  # Initialize with random values

# Define the cost function (Mean Squared Error)
def compute_cost(X, y, theta):
    predictions = np.dot(X, theta)
    mse = np.mean((predictions - y.reshape(-1, 1)) ** 2)
    return mse

# Stochastic Gradient Descent function to update parameters
def stochastic_gradient_descent(X, y, theta, learning_rate, epochs):
    m = len(y)
    costs = []
    for epoch in range(epochs):
        for i in range(m):
            random_index = np.random.randint(m)
            X_i = X[random_index:random_index+1]
            y_i = y[random_index:random_index+1]
            
            # Compute prediction and error for the single instance
            prediction = np.dot(X_i, theta)
            error = prediction - y_i
            
            # Compute gradient for the single instance
            gradient = 2 * X_i.T.dot(error)
            
            # Update parameters
            theta -= learning_rate * gradient
            
        # Compute and store cost for the epoch
        cost = compute_cost(X, y, theta)
        costs.append(cost)
        
        # Print progress
        if epoch % 100 == 0:
            print(f"Epoch {epoch}: Cost = {cost}")
    
    return theta, costs

# Perform Stochastic Gradient Descent
theta_best, costs = stochastic_gradient_descent(X_b, y, theta, learning_rate, epochs)

# Print final parameters and cost
print("\nFinal Parameters (theta):")
print(theta_best)
print(f"Final Cost: {costs[-1]}")

# Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, X_b.dot(theta_best), color='red', label='Best Fit Line')
plt.title('Best Fit Line using Stochastic Gradient Descent')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.legend()
plt.grid(True)
plt.show()
```
